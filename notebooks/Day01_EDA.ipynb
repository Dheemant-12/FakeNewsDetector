```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
fake_df = pd.read_csv("../data/Fake.csv")
true_df = pd.read_csv("../data/True.csv")

# Add labels
fake_df["label"] = 0
true_df["label"] = 1

# Combine datasets
df = pd.concat([fake_df, true_df], ignore_index=True)

print("=" * 50)
print("Dataset Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nFirst Five Rows:")
print(df.head())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nClass Distribution:")
print(df["label"].value_counts())

# Text length analysis
df["text_length"] = df["text"].apply(len)

print("\nAverage Text Length:")
print(df["text_length"].mean())

# Plot class distribution
plt.figure(figsize=(5,4))
df["label"].value_counts().plot(kind="bar")
plt.xticks([0,1], ["Fake","Real"], rotation=0)
plt.title("Class Distribution")
plt.xlabel("News Type")
plt.ylabel("Count")
plt.show()

# Plot text length histogram
plt.figure(figsize=(8,5))
plt.hist(df["text_length"], bins=50)
plt.title("Distribution of Article Lengths")
plt.xlabel("Characters")
plt.ylabel("Frequency")
plt.show()
```

```python
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt

# -----------------------------
# Duplicate Analysis
# -----------------------------
print("="*50)
print("Duplicate Rows:")
print(df.duplicated().sum())

print("\nDuplicate News Articles:")
print(df.duplicated(subset=["text"]).sum())

# -----------------------------
# Subject Distribution
# -----------------------------
print("\nSubjects:")
print(df["subject"].value_counts())

plt.figure(figsize=(8,5))
df["subject"].value_counts().plot(kind="bar")
plt.title("News Categories")
plt.xlabel("Subject")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -----------------------------
# Publication Dates
# -----------------------------
df["date"] = pd.to_datetime(df["date"], errors="coerce")

print("\nDate Range:")
print(df["date"].min())
print(df["date"].max())

# -----------------------------
# Word Clouds
# -----------------------------
fake_text = " ".join(df[df["label"] == 0]["text"].astype(str))
real_text = " ".join(df[df["label"] == 1]["text"].astype(str))

fake_cloud = WordCloud(
    width=800,
    height=400,
    background_color="white"
).generate(fake_text)

real_cloud = WordCloud(
    width=800,
    height=400,
    background_color="white"
).generate(real_text)

plt.figure(figsize=(12,5))
plt.imshow(fake_cloud)
plt.axis("off")
plt.title("Most Frequent Words in Fake News")
plt.show()

plt.figure(figsize=(12,5))
plt.imshow(real_cloud)
plt.axis("off")
plt.title("Most Frequent Words in Real News")
plt.show()

# -----------------------------
# Top 20 Most Frequent Words
# -----------------------------
words = fake_text.lower().split()

counter = Counter(words)

print("\nTop 20 Most Common Words (Fake News):")

for word, count in counter.most_common(20):
    print(word, count)
```
